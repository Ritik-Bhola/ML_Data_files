```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor(random_state=42)

from sklearn.linear_model import LogisticRegression
log=LogisticRegression()

from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()

#Simple Imputer
from sklearn.impute import IterativeImputer

#Knn Imputer
from sklearn.impute import KNNImputer


import warnings
warnings.filterwarnings("ignore")
```

```python
data={"SibSp":[1,1,0,1,0,0],"Fare":[7.2500,71.2833,7.2950,53.1000,8.0500,8.4583],"Age":[22,38,26,35,35,np.NaN]}
data

X=pd.DataFrame(data)
X
```

# Iterative Imputer

Iterative Imputer is simply a Machine Learning Model that simply Imputes the missing values with predicted values. The approach is simple and easy to understand although it lack the data understanding that has curbed by KNN Imputer

```python
impute=IterativeImputer(lr)
impute.fit_transform(X)
```

```python
impute=IterativeImputer(rfr)
impute.fit_transform(X)
```
